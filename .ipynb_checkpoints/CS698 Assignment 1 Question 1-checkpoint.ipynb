Part 3(a) : PureExploitation strategy

```python
def PureExploitation(env, maxEpisodes=1000, print_table=False, ret_reward_list=False, ret_action_list=False):
  q = np.zeros(env.action_space.n)
  N = np.zeros(env.action_space.n)
  episodes=0
  q_book_keeping = np.zeros((maxEpisodes+1,env.action_space.n))
  reward_list=[]
  action_list=[]

  while(episodes<maxEpisodes):
    a = np.argmax(q)
    next_state, reward, done = env.step(a)
    N[a]+=1
    q[a]+= (reward - q[a])*1.0/N[a]
    episodes+=1
    q_book_keeping[episodes]=q
    reward_list.append(reward)
    action_list.append(a)
    if(done):
      _ = env.reset()
    if(print_table):
      print("Action {}, Reward {} -> {}".format(a,reward,q))
  
  if(ret_reward_list):
    if(ret_action_list):
      return q_book_keeping,q, reward_list, action_list
    else :
      return q_book_keeping,q, reward_list

  return q_book_keeping,q

def PureExploitation_Optimistic_initialization(env, maxEpisodes=1000, print_table=False):
  q = np.zeros(env.action_space.n)+5
  N = np.zeros(env.action_space.n)
  episodes=0
  q_book_keeping = np.zeros((maxEpisodes+1,env.action_space.n))

  while(episodes<maxEpisodes):
    a = np.argmax(q)
    next_state, reward, done = env.step(a)
    N[a]+=1
    q[a]+= (reward - q[a])*1.0/N[a]
    episodes+=1
    q_book_keeping[episodes]=q
    if(done):
      _ = env.reset()
    if(print_table):
      print("Action {}, Reward {} -> {}".format(a,reward,q))
  
  return q_book_keeping,q

#Part 3(a) : run PureExploitation strategy on bernoulli bandit environment
print("Running PureExploitation strategy on bernoulli bandit environment")
env = gym.make("gym_foo:Bernoulli_Bandit-v0")
env.__init__(0.6,0.9, set_seed = seed)
q_book_keeping, q = PureExploitation(env,5,print_table=True)
```

Part 3(b) : PureExploration strategy

```python
def PureExploration(env, maxEpisodes=1000, print_table=False, ret_reward_list=False, ret_action_list=False):
  q = np.zeros(env.action_space.n)
  N = np.zeros(env.action_space.n)
  episodes=0
  q_book_keeping = np.zeros((maxEpisodes+1,env.action_space.n))
  reward_list=[]
  action_list=[]

  while(episodes<maxEpisodes):
    a = env.action_space.sample()
    next_state, reward, done = env.step(a)
    N[a]+=1
    q[a]+= (reward - q[a])*1.0/N[a]
    episodes+=1
    q_book_keeping[episodes]=q
    reward_list.append(reward)
    action_list.append(a)
    if(done):
      _ = env.reset()
    if(print_table):
      print("Action {}, Reward {} -> {}".format(a,reward,q))
  
  if(ret_reward_list):
    if(ret_action_list):
      return q_book_keeping,q, reward_list, action_list
    else :
      return q_book_keeping,q, reward_list
  
  return q_book_keeping,q

#Part 3(b) : run PureExploration strategy on bernoulli bandit environment
print("Running PureExploration strategy on bernoulli bandit environment")
env = gym.make("gym_foo:Bernoulli_Bandit-v0")
env.__init__(0.5,0.5, set_seed = seed)
q_book_keeping, q = PureExploration(env,maxEpisodes=5,print_table=True)
```

Part 3(c) : ε-Greedy strategy

```python
def eps_greedy(env, maxEpisodes=1000, eps = 0.3, ret_reward_list=False, ret_action_list=False):
  q = np.zeros(env.action_space.n)
  N = np.zeros(env.action_space.n)
  episodes=0
  q_book_keeping = np.zeros((maxEpisodes+1,env.action_space.n))
  reward_list=[]
  action_list=[]

  while(episodes<maxEpisodes):
    # a = env.action_space.sample()
    if(np.random.uniform()<eps):  #explore
      a = env.action_space.sample()
    else :  #exploit
      a = np.argmax(q)

    next_state, reward, done = env.step(a)
    N[a]+=1
    q[a]+= (reward - q[a])*1.0/N[a]
    episodes+=1
    q_book_keeping[episodes]=q
    reward_list.append(reward)
    action_list.append(a)
    if(done):
      _ = env.reset()
  
  if(ret_reward_list):
    if(ret_action_list):
      return q_book_keeping,q, reward_list, action_list
    else :
      return q_book_keeping,q, reward_list
  
  return q_book_keeping,q

env = gym.make("gym_foo:Bernoulli_Bandit-v0")
env.__init__(0.6,0.9, set_seed = seed)

for eps in np.linspace(0.0,1.0,5):
  _,q = eps_greedy(env, 1000, eps)
  print("Q values for {:0.2f}-greedy strategy {}".format(eps,q))
```

Part 3(d) : decaying ε-Greedy strategy

```python
def decaying_eps_greedy(env, maxEpisodes=1000, decay_type = 'exponential', decay_rate = 1.0, ret_reward_list=False, ret_action_list=False):
  q = np.zeros(env.action_space.n)
  N = np.zeros(env.action_space.n)
  episodes=0
  q_book_keeping = np.zeros((maxEpisodes+1,env.action_space.n))
  reward_list=[]
  action_list=[]

  while(episodes<maxEpisodes):
    if(decay_type == 'exponential'):
      eps = np.exp(-1*decay_rate*episodes)
    elif(decay_type == 'linear'):
      eps = max(0, 1-decay_rate*episodes)
    # print("Eps for episode {} is {:0.3f}".format(episodes,eps))

    if(np.random.uniform()<eps):  #explore
      a = env.action_space.sample()
    else :  #exploit
      a = np.argmax(q)

    next_state, reward, done = env.step(a)
    N[a]+=1
    q[a]+= (reward - q[a])*1.0/N[a]
    episodes+=1
    q_book_keeping[episodes]=q
    reward_list.append(reward)
    action_list.append(a)
    if(done):
      _ = env.reset()

  if(ret_reward_list):
    if(ret_action_list):
      return q_book_keeping,q, reward_list, action_list
    else :
      return q_book_keeping,q, reward_list
  
  return q_book_keeping,q
```

Part 3(e) : Softmax strategy

```python
def Softmax_strategy(env, maxEpisodes=1000, temperature = 1.0, ret_reward_list=False, ret_action_list=False):
  q = np.zeros(env.action_space.n)
  N = np.zeros(env.action_space.n)
  episodes=0
  q_book_keeping = np.zeros((maxEpisodes+1,env.action_space.n))
  reward_list=[]
  action_list=[]
  # print("Debug : q shape {}, N shape {}, q_book shape {}".format(q.shape,N.shape,q_book_keeping.shape))

  while(episodes<maxEpisodes):
    to_softmax = q/temperature - np.mean(q/temperature) #to prevent numerical overflow
    probs = np.exp(to_softmax)/sum(np.exp(to_softmax))
    a = np.random.choice(env.action_space.n, 1, p = probs)[0]

    next_state, reward, done = env.step(a)
    N[a]+=1
    q[a]+= (reward - q[a])*1.0/N[a]
    episodes+=1
    q_book_keeping[episodes] = q
    reward_list.append(reward)
    action_list.append(a)
    if(done):
      _ = env.reset()
  
  if(ret_reward_list):
    if(ret_action_list):
      return q_book_keeping,q, reward_list, action_list
    else :
      return q_book_keeping,q, reward_list

  return q_book_keeping,q
```

Part 3(f) : UCB strategy

```python
def UCB_strategy(env, maxEpisodes=1000, c=0.1, ret_reward_list=False, ret_action_list=False):
  q = np.zeros(env.action_space.n)
  N = np.zeros(env.action_space.n)
  episodes=0
  q_book_keeping = np.zeros((maxEpisodes+1,env.action_space.n))
  reward_list=[]
  action_list=[]

  while(episodes<maxEpisodes):
    if(episodes<env.action_space.n):
      a = episodes
    else:
      U = c * np.sqrt(np.log(episodes)/N)
      a = np.argmax(q + U)

    next_state, reward, done = env.step(a)
    N[a]+=1
    q[a]+= (reward - q[a])*1.0/N[a]
    episodes+=1
    q_book_keeping[episodes]=q
    reward_list.append(reward)
    action_list.append(a)
    if(done):
      _ = env.reset()
  
  if(ret_reward_list):
    if(ret_action_list):
      return q_book_keeping,q, reward_list, action_list
    else :
      return q_book_keeping,q, reward_list

  return q_book_keeping,q
```
